# MODEL SETUP AND DOWNLOAD

```python
#@markdown ### Model Setup and Download
#@markdown Select the models to download:
download_stable_diffusion = True #@param {type:"boolean"}
download_AnimateDiff = True #@param {type:"boolean"}

#@markdown **Download Settings:**
use_aria2 = True #@param {type:"boolean"} {"tooltip":"Use aria2 for faster downloads"}
max_connections = 16 #@param {type:"slider", min:1, max:32, step:1}
min_split_size = "1M" #@param ["1M", "5M", "10M", "20M"]

#@markdown **Advanced Settings:**
auto_retry = False #@param {type:"boolean"} {"tooltip":"Automatically retry failed downloads"}
retry_count = 3 #@param {type:"slider", min:1, max:10, step:1}

import os

# Function to execute model setup and downloads
def setup_and_download():
    os.system("cd /content")
    os.system("git clone https://github.com/Artiprocher/DiffSynth-Studio")
    os.system("cd /content/DiffSynth-Studio")
    os.system("pip install -q einops transformers controlnet-aux==0.0.7 sentencepiece imageio imageio-ffmpeg")
    os.system("apt -y install -qq aria2")

    # Dictionary of model URLs and file names
    models_info = {
        "stable_diffusion": {
            "url": "https://civitai.com/api/download/models/229575",
            "output": "aingdiffusion_v12.safetensors"
        },
        "AnimateDiff": {
            "url": "https://huggingface.co/guoyww/animatediff/resolve/main/mm_sd_v15_v2.ckpt",
            "output": "mm_sd_v15_v2.ckpt"
        },
        # Add more models here as needed
    }

    # Download selected models
    def download_model(model_key):
        if model_key in models_info:
            info = models_info[model_key]
            full_path = f"/content/DiffSynth-Studio/models/{model_key}"
            os.makedirs(full_path, exist_ok=True)
            download_command = f"aria2c --console-log-level=error -c -x {max_connections} -s {max_connections} -k {min_split_size} {info['url']} -d {full_path} -o {info['output']}"
            if not use_aria2:
                download_command = f"wget {info['url']} -O {full_path}/{info['output']}"
            os.system(download_command)
            print(f"Downloaded {model_key} model.")

    if download_stable_diffusion:
        download_model("stable_diffusion")

    if download_AnimateDiff:
        download_model("AnimateDiff")

    if auto_retry:
        for _ in range(retry_count):
            if download_stable_diffusion:
                download_model("stable_diffusion")
            if download_AnimateDiff:
                download_model("AnimateDiff")

    print("Selected models downloaded and installed.")

# Run the function to setup and download models
setup_and_download()
```

```python
#@markdown ### Model Download Configuration
model_url_1 = "https://huggingface.co/Disty0/SoteMix/resolve/main/SoteMix.safetensors?download=true" #@param {type:"string"}
target_directory_1 = "/content/DiffSynth-Studio/models/stable_diffusion" #@param {type:"string"}
file_name_1 = "SoteMix.safetensors" #@param {type:"string"}

#@markdown ### Model Download Configuration for "lora"
model_url_2 = "https://huggingface.co/lraza8632/LoRAs/resolve/d76e87a8016fe1bda531312ec1a45d0e2bb0e38d/Animesh-Lora.safetensors?download=true" #@param {type:"string"}
target_directory_2 = "/content/DiffSynth-Studio/models/lora" #@param {type:"string"}
file_name_2 = "AniMesh.safetensors" #@param {type:"string"}

import os
import requests
from tqdm.notebook import tqdm

# Function to download the model
def download_model(url, output_path):
    response = requests.get(url, stream=True)
    response.raise_for_status()

    total_size = int(response.headers.get('content-length', 0))

    with open(output_path, 'wb') as f:
        for chunk in tqdm(response.iter_content(chunk_size=8192), total=total_size//8192, unit='KB', unit_scale=True):
            if chunk:
                f.write(chunk)

    print(f"Model downloaded and saved to {output_path}")

# Define the full target paths for the files
target_file_1 = os.path.join(target_directory_1, file_name_1)
target_file_2 = os.path.join(target_directory_2, file_name_2)

# Create the directories if they do not exist
os.makedirs(target_directory_1, exist_ok=True)
os.makedirs(target_directory_2, exist_ok=True)

# Download the models
download_model(model_url_1, target_file_1)
download_model(model_url_2, target_file_2)

#@markdown ### Video Input Configuration / settings
video_file = "/content/input_video.mp4" #@param {type:"string"}
output_folder = "/content/toon_video" #@param {type:"string"}
start_frame_id = 0 #@param {type:"integer"}
end_frame_id = 30 #@param {type:"integer"}
frame_height = 1024 #@param {type:"integer"}
frame_width = 1024 #@param {type:"integer"}
fps = 25 #@param {type:"integer"}

#@markdown ### Pipeline Configuration
seed = 0 #@param {type:"integer"}
prompt = "animated cartoon clay character, character cartoon made of clay, clay material, clay, clay toon, 3D clay, animated" #@param {type:"string"}
negative_prompt = "verybadimagenegative_v1.3" #@param {type:"string"}
cfg_scale = 8.0 #@param {type:"number"}
clip_skip = 2 #@param {type:"slider", min:0, max:10, step:1}
denoising_strength = 1.0 #@param {type:"number"}
num_inference_steps = 20 #@param {type:"slider", min:5, max:50, step:1}
animatediff_batch_size = 16 #@param {type:"integer"}
animatediff_stride = 10 #@param {type:"integer"}
unet_batch_size = 2 #@param {type:"integer"}
controlnet_batch_size = 2 #@param {type:"integer"}
cross_frame_attention = False #@param {type:"boolean"}

from diffsynth import SDVideoPipelineRunner

# Load the base configuration
config_stage_2_template = {
    "models": {
        "model_list": [
            target_file_1,
            "models/AnimateDiff/mm_sd_v15_v2.ckpt",
            "models/ControlNet/control_v11f1e_sd15_tile.pth",
            "models/ControlNet/control_v11p_sd15_lineart.pth"
        ],
        "textual_inversion_folder": "models/textual_inversion",
        "device": "cuda",
        "lora_alphas": [
            {
                "model": target_file_2,
                "scale": 2.0
            },
        ],
        "controlnet_units": [
            {
                "processor_id": "tile",
                "model_path": "models/ControlNet/control_v11f1e_sd15_tile.pth",
                "scale": 1.0
            },
            {
                "processor_id": "lineart",
                "model_path": "models/ControlNet/control_v11p_sd15_lineart.pth",
                "scale": 1.0
            }
        ]
    },
    "data": {
        "input_frames": {
            "video_file": video_file,
            "image_folder": None,
            "height": frame_height,
            "width": frame_width,
            "start_frame_id": start_frame_id,
            "end_frame_id": end_frame_id
        },
        "controlnet_frames": [
            {
                "video_file": video_file,
                "image_folder": None,
                "height": frame_height,
                "width": frame_width,
                "start_frame_id": start_frame_id,
                "end_frame_id": end_frame_id
            },
            {
                "video_file": video_file,
                "image_folder": None,
                "height": frame_height,
                "width": frame_width,
                "start_frame_id": start_frame_id,
                "end_frame_id": end_frame_id
            }
        ],
        "output_folder": output_folder,
        "fps": fps
    },
    "pipeline": {
        "seed": seed,
        "pipeline_inputs": {
            "prompt": prompt,
            "negative_prompt": negative_prompt,
            "cfg_scale": cfg_scale,
            "clip_skip": clip_skip,
            "denoising_strength": denoising_strength,
            "num_inference_steps": num_inference_steps,
            "animatediff_batch_size": animatediff_batch_size,
            "animatediff_stride": animatediff_stride,
            "unet_batch_size": unet_batch_size,
            "controlnet_batch_size": controlnet_batch_size,
            "cross_frame_attention": cross_frame_attention,
            # The following parameters will be overwritten. You don't need to modify them.
            "input_frames": [],
            "num_frames": 30,
            "width": frame_width,
            "height": frame_height,
            "controlnet_frames": []
        }
    }
}

# Create and run the video pipeline runner
runner = SDVideoPipelineRunner()
runner.run(config_stage_2_template)

#@markdown ### Display Processed Video
video_path = output_folder + "/video.mp4" #@param {type:"string"}

import moviepy.editor as mp

# Display the video in the notebook
video_clip = mp.VideoFileClip(video_path)
video_clip.ipython_display(fps=fps, maxduration=240)
```
