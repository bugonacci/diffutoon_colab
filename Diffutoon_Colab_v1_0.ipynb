# MODEL SETUP AND DOWNLOAD

```python
#@markdown ### Model Setup and Download
#@markdown Select the models to download:
download_stable_diffusion = True #@param {type:"boolean"}
download_AnimateDiff = True #@param {type:"boolean"}

#@markdown **Download Settings:**
use_aria2 = True #@param {type:"boolean"} {"tooltip":"Use aria2 for faster downloads"}
max_connections = 16 #@param {type:"slider", min:1, max:32, step:1}
min_split_size = "1M" #@param ["1M", "5M", "10M", "20M"]

#@markdown **Advanced Settings:**
auto_retry = False #@param {type:"boolean"} {"tooltip":"Automatically retry failed downloads"}
retry_count = 3 #@param {type:"slider", min:1, max:10, step:1}

import os

# Function to execute model setup and downloads
def setup_and_download():
    os.system("cd /content")
    os.system("git clone https://github.com/Artiprocher/DiffSynth-Studio")
    os.system("cd /content/DiffSynth-Studio")
    os.system("pip install -q einops transformers controlnet-aux==0.0.7 sentencepiece imageio imageio-ffmpeg")
    os.system("apt -y install -qq aria2")

    # Dictionary of model URLs and file names
    models_info = {
        "stable_diffusion": {
            "url": "https://civitai.com/api/download/models/229575",
            "output": "aingdiffusion_v12.safetensors"
        },
        "AnimateDiff": {
            "url": "https://huggingface.co/guoyww/animatediff/resolve/main/mm_sd_v15_v2.ckpt",
            "output": "mm_sd_v15_v2.ckpt"
        },
        # Add more models here as needed
    }

    # Download selected models
    def download_model(model_key):
        if model_key in models_info:
            info = models_info[model_key]
            full_path = f"/content/DiffSynth-Studio/models/{model_key}"
            os.makedirs(full_path, exist_ok=True)
            download_command = f"aria2c --console-log-level=error -c -x {max_connections} -s {max_connections} -k {min_split_size} {info['url']} -d {full_path} -o {info['output']}"
            if not use_aria2:
                download_command = f"wget {info['url']} -O {full_path}/{info['output']}"
            os.system(download_command)
            print(f"Downloaded {model_key} model.")

    if download_stable_diffusion:
        download_model("stable_diffusion")

    if download_AnimateDiff:
        download_model("AnimateDiff")

    if auto_retry:
        for _ in range(retry_count):
            if download_stable_diffusion:
                download_model("stable_diffusion")
            if download_AnimateDiff:
                download_model("AnimateDiff")

    print("Selected models downloaded and installed.")

# Run the function to setup and download models
setup_and_download()
```

```python
#@markdown ### Model Download Configuration
model_url = "https://civitai.com/api/download/models/224732?type=Model&format=SafeTensor&size=pruned&fp=fp16" #@param {type:"string"}
target_directory = "/content/DiffSynth-Studio/models/stable_diffusion" #@param {type:"string"}
file_name = "AniMesh.safetensors" #@param {type:"string"}

import os
import requests
from tqdm.notebook import tqdm

# Define the full target path for the file
target_file = os.path.join(target_directory, file_name)

# Create the directory if it does not exist
os.makedirs(target_directory, exist_ok=True)

# Function to download the model
def download_model(url, output_path):
    # Stream download with progress bar
    response = requests.get(url, stream=True)
    response.raise_for_status()  # Check that the request was successful

    # Get the total file size from the response headers
    total_size = int(response.headers.get('content-length', 0))

    # Write the file to the target location with progress bar
    with open(output_path, 'wb') as f:
        for chunk in tqdm(response.iter_content(chunk_size=8192), total=total_size//8192, unit='KB', unit_scale=True):
            if chunk:  # Filter out keep-alive new chunks
                f.write(chunk)

    print(f"Model downloaded and saved to {output_path}")

# Call the download function
download_model(model_url, target_file)
```
